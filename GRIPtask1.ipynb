# Importing library files

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
```

```python
%matplotlib inline
```

# Importing the data

```python
data=pd.read_csv("http://bit.ly/w-data")
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

# Splitting the dataset

```python
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=100)
```

# Making a linear regression model

```python
model=linear_model.LinearRegression()
```

# Fitting the modle with the dataset

```python
model.fit(X_train,Y_train)
```

```python
ypred=model.predict(x_test)
```

# Equation of Linear regression model with the train data

```python
y=model.coef_*(X_train)+model.intercept_
```

# Plot of the training data

```python
plt.scatter(X_train,Y_train,color="black")
plt.plot(X_train,y,color="blue")
```

# Plot of the test data and predicted data

```python
plt.scatter(x_test,y_test,color="black")
plt.plot(x_test,ypred,color="blue")
plt.xlabel("Hours")
plt.ylabel("Scores")
```

```python
from sklearn import metrics
x=metrics.mean_absolute_error(ypred,y_test)
print("The Mean absolute error of this Model :"+str(x))
```
